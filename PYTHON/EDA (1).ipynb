```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Clean genre column
# If a movie has multiple genres like "Action|Adventure", pick the first as main genre
df['Main_Genre'] = df['genres'].str.strip().str.split('|').str[0]

# Step 2: Count movies per genre
genre_counts = df['Main_Genre'].value_counts()

# Step 3: Horizontal bar chart
plt.figure(figsize=(12,8))
bars = plt.barh(genre_counts.index, genre_counts.values, color='skyblue')

# Step 4: Add numbers on each bar
for i, value in enumerate(genre_counts.values):
    plt.text(value + 1, i, str(value), va='center')

# Step 5: Labels and title
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.title("Number of Movies Released per Genre")
plt.gca().invert_yaxis()  # largest bars on top
plt.tight_layout()
plt.show()
```
